```python
from fbprophet import Prophet
```

```python
msft['ds'] = msft.index
msft['y'] = msft['Adj_Close']

forecast_data = msft[['ds', 'y']].copy()
forecast_data.reset_index(inplace=True)
forecast_data.head()
```

```python
m = Prophet()
m.fit(msft);

future = m.make_future_dataframe(periods=200, freq='H')

forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig = m.plot(forecast, xlabel='Data', ylabel='Price (USD)');
plt.title('Microsoft Adjusted Closing Prices (USD)')
fig.gca().yaxis.set_major_formatter(matplotlib.ticker.StrMethodFormatter('${x:,.0f}'))
```

```python
m.plot_components(forecast)
```
